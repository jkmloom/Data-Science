# Reading and Writing Files

```python
text_content = """
example1
example2
example3
"""
```

```python
# Write in a text file
with open('example.txt', 'w') as file:
    file.write(text_content)
```

```python
# read a text file
with open('example.txt', 'r') as file:
    data = file.read()
print(data)
```

```python
# csv >> comma separated values
data = [["Name", "Course", "Fee"],
       ["Jatin", "DS", "20000"],
       ["Kaushik", "Astro Physics", "15000"]]
```

```python
for i in data:
    print(i)
```

```python
import csv
# writing data in a csv file
with open('example_csv.csv', 'w') as f:
    w = csv.writer(f)
    for i in data:
        w.writerow(i)
```

```python
with open("example_csv.csv", 'r') as f:
    r = csv.reader(f)
    for i in r:
        print(i)
```

```python
# JSON >> JavaScript Object Notation
# data stored in key:value pair
data = {"name": "Ajay",
       "course": "Data Science",
       "fee": 20000}
data
```

```python
type(data)
```

```python
# make JSON file out of dict
import json
# dump data in a json file: file_json.json
with open("file_json.json", 'w') as f:
    json.dump(data, f) # syntax: json.dump(what, where)
```

```python
with open("file_json.json", 'r') as f:
    data = json.load(f)
print(data)
```

JSON files are very useful when working with APIs

```python
# writing binary data in a file
with open("test_bin.bin", 'wb') as f:
    # .bin files are binary files
    # 'wb' >> write byte
    f.write(b"\x48\x65\x6C\x6C\x6F\x20\x57\x6F\x72\x6C\x64")
```

```python
# reading binary data from a file
with open("test_bin.bin", 'rb') as f:
    # 'rb' >> read byte
    print(f.read())
```

```python
# buffered reading and writing
# it allows the program to read and write data in chunks : generally used for huge data
import io
# io >> input output
with open("test_buf.txt", 'wb') as f:
    file = io.BufferedWriter(f)
    file.write(b"Data modulation is the process of converting digital data signals into analog signals suitable for transmission over a communication medium, such as by varying properties of a carrier wave like amplitude, frequency, or phase.")
    file.write(b"Hello World, how are you?")
    file.write(b"I am working on data science")
    file.flush()
```

```python
with open("test_buf.txt", 'rb') as f:
    file = io.BufferedReader(f)
    data = file.read(100) # read the first chunk of 100 bytes
    print(data)
```
