```python
open('file.txt', 'r')
# if file do not exist then in 'r' mode >> error!

# r >> read mode
# w >> write mode
# a >> append mode
# r+ >> reading and writing mode
```

```python
file_name = 'cartoon.txt'
file = open(file_name, 'w')
# if file do not exist then in 'w' mode >> create file and read/write it
```

```python
file.write("Hello this is my first line in this text file\n")
# o/p: memory (bytes) written in the file
```

```python
# if you open the file now, you won't see any changes
# to see the changes the file muct be closed first
file.close()
```

```python
file # file attributes
```

```python
# file close status
file.closed
```

```python
file = open(file_name, 'r')
file.closed
```

```python
file.close()
file.closed # for file closed >> return value = True
```

```python
# last status of mode the file was opened in 
file.mode # 'r' >> read mode
```

```python
# can not write any data in a closed file
file = open(file_name, 'w')
file.write("This is the second line in this file name >> cartoon.txt")
# it will truncate (delete) all the older data and overwrite: cartoon.txt
```

```python
file.close()
file.closed
```

```python
# writing multiple lines
file = open(file_name, 'w')
file.write("This is the first line\n")
file.write("This is the second line\n")
file.write("This is the third line\n")
file.write("This is the fourth line")
file.close()
# writing anything after the above line will delete the current written data
```

```python
file.closed
```

```python
# Appending data in a file
# 'a' >> append mode
file = open(file_name, 'a')
file.write("\nThis is the fifth line")
# this will appen a new line without affecting the older data
file.close()
file.closed
```

```python
file = open(file_name, 'r')
# since the file is opened in read mode, the contents could be read
```

```python
for i in file:
    print(i)
```

```python
# can use file_name.read() for reading a file
file.read() # the cursor by default is at the last line
```

```python
print(file.read()) # same output >> nothing
```

```python
file.tell() # tells the current position of the cursor
```

```python
# for reading the file from the 0th position
file.seek(0) # bring the cursor at 0th position
```

```python
file.close()
file.closed
```

```python
f = open(file_name, 'r')
print(f.readline())
f.close()
```

```python
f = open(file_name, 'r')
f.seek(0) # bring the cursor to the 0th position
print(f.read()) # will read everything from the start
f.close()
f.closed
```

```python
# calling the file data in a list/array
f = open(file_name, 'r')
print(f.readlines()) # all the lines will be the element of a list
f.close()
```

```python
# Accessing file data as list/array
f = open(file_name, 'r')
data = f.readlines()
data[2]
```

```python
f.close()
f.closed
```

```python
# read the file but without manually closing it
with open('file.txt', 'w') as f:
    f.write("This is the first line")
    # 'with' statement ensures that the file is properly closed after execution of code
    # important >> free up system resources and avoid issues like data corruption of file lock
```

```python
with open('file.txt', 'r') as f:
    line = f.readlines()
    print(line)
```
